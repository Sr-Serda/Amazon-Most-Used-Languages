```python
import requests
from io import BytesIO
from PIL import Image

def get_user_image(user_name, img_size_width, img_size_height):
    url = f'https://api.github.com/users/{user_name}'
    response = requests.get(url)

    user_data = response.json()

    avatar_url = user_data.get('avatar_url')

    avatar_response = requests.get(avatar_url)
    img = Image.open(BytesIO(avatar_response.content))
    img.thumbnail([img_size_width, img_size_height])
    img.show()
    
```

Levantando todos os repositórios

```python
repos_list = []
page_num = 1

while True:
     try:
         url_page = f'{url}?page={page_num}'
         response = requests.get(url_page, headers=headers)
        
         # # Levantar uma exceção se houver erro no status da resposta
         response.raise_for_status()
        
         data = response.json()
        
         # # Verificar se a resposta está vazia (fim da paginação)
         if not data:
              break
        
         repos_list.append(data)  # Adicionar os resultados na lista
         page_num += 1
     except Exception as e:
         print(f"Erro na página {page_num}: {e}")
         break

```

Resgatando as linguagens dos repositórios

```python
languages_repo = []

for list in repos_list:
    for sublist in list:
        found_language = sublist.get('language')
        
        if found_language is None:
            found_language = ''
        
        languages_repo.append(found_language)
```

```python
languages_repo
```

Resgatando os nomes dos repositórios

```python
name_repo = []

for list in repos_list:
    for sublist in list:
        found_name = sublist.get('name')
        
        if found_name is None:
            found_name = ''
        
        name_repo.append(found_name)
```

```python
name_repo
```

Transformando Dicionário em lista

```python
repo_name_language = dict(zip(name_repo, languages_repo))
```

```python
new_repo_name_language = {key: [key, value] for key, value in repo_name_language.items()}
new_repo_name_language
```

```python
list_repo_name_language = [*repo_name_language.items()]
```

```python
list_repo_name_language
```

Transformando dicionário em dataframe

```python
dataframe_repo_name_language = pd.DataFrame.from_dict(repo_name_language, orient='index')
```

```python
dataframe_repo_name_language.reset_index(inplace=True)
dataframe_repo_name_language.columns = ['Repo', 'Language']
```

```python
dataframe_repo_name_language.to_csv('amzn.csv')
dataframe_repo_name_language
```
